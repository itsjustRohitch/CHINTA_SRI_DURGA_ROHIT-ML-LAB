```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

```python
zeros_arr = np.zeros((3, 4))   # 3 rows, 4 columns
print(zeros_arr)
```

```python
ones_arr = np.ones((2, 3))     # 2 rows, 3 columns
print(ones_arr)
```

```python
print(arr.shape)        # (5,)
print(zeros_arr.shape) # (3, 4)
```

```python
reshaped = arr.reshape((5, 1))   # 5 rows, 1 column
print(reshaped)
print(reshaped.shape)
```
