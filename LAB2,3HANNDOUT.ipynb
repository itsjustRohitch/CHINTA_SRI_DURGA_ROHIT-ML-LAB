```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("USA_Housing.csv")
```

```python
df.isnull()               # Detect missing values (True/False)
```

```python
df.isnull().sum()         # Count missing values per column
```

```python
df.notnull()              # Detect non-missing values
```

```python
df.dropna()               # Drop rows with any missing values
df.dropna(axis=1)         # Drop columns with missing values
```

```python
df.fillna(0)              # Fill missing values with 0
df.fillna(df.mean(numeric_only=True))  # Fill missing numerical values with mean
```

```python
df["Avg. Area Income"] = df["Avg. Area Income"].astype(float)
df["Avg. Area House Age"] = df["Avg. Area House Age"].astype(float)
df["Avg. Area Number of Rooms"] = df["Avg. Area Number of Rooms"].astype(float)
df["Avg. Area Number of Bedrooms"] = df["Avg. Area Number of Bedrooms"].astype(float)
df["Area Population"] = df["Area Population"].astype(float)
df["Price"] = df["Price"].astype(float)
```

```python
df.rename(columns={
    "Avg. Area Income": "Avg_Area_Income",
    "Avg. Area House Age": "Avg_Area_House_Age",
    "Avg. Area Number of Rooms": "Avg_Area_Num_Rooms",
    "Avg. Area Number of Bedrooms": "Avg_Area_Num_Bedrooms",
    "Area Population": "Area_Population"
}, inplace=True)
```

```python
df["Address"].replace("New", "NEW")
```

```python
# Line plot: Income vs Price
df["Price"].plot(kind="line")
plt.title("House Price Trend")
plt.ylabel("Price")
plt.show()
```

```python
# Bar plot: Average price by house age (binned)
df.groupby("Avg_Area_House_Age")["Price"].mean().head(10).plot(kind="bar")
plt.title("Average Price by House Age")
plt.show()
```

```python
# Scatter plot: Income vs Price
plt.scatter(df["Avg_Area_Income"], df["Price"])
plt.xlabel("Average Area Income")
plt.ylabel("House Price")
plt.title("Income vs House Price")
plt.show()
```

```python
# Histogram: House prices
df["Price"].plot(kind="hist", bins=15)
plt.title("Distribution of House Prices")
plt.show()
```

```python
# Box plot: House prices
df["Price"].plot(kind="box")
plt.title("Box Plot of House Prices")
plt.show()
```
